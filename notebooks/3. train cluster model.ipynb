```python
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor 
```

```python
df = pd.read_csv("dataframe-auditoria-small-with-anomalies.csv")
df.drop(columns=['Unnamed: 0'],inplace=True)
print(df.shape)
df.head()
```

```python
clf = LocalOutlierFactor(novelty=True,contamination=0.07)
clf = clf.fit(df)
```

```python
y_pred = clf.predict(df)
y_pred = pd.Series(y_pred)
df['cluster'] = y_pred
```
